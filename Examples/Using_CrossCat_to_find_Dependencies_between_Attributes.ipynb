**Goal**: The goal of this use case is to show a user how to use CrossCat to identify dependencies between attributes.
    
**Need**: Very often, the first step an analyst takes when presented with a large database is to attempt to find attributes that are related. This is often accomplished using pair-wise scatter plots or some other means of visualizing attributes plotted "against" each other. However, in a high-dimensional database this approach could result in a prohibitively large number attribute pairs (or other tuples) to be investigated. Moreover, this approach does not reliably capture higher order relationships between multiple attributes. In order to reduce the burden on the analysts and identify relationships that cannot be easily visualized, reliable, automated techniques such as CrossCat must be used to first reduce the search space for analysis.

**Data Set Used for This Example**: In this example, we have used a subset of a listing of flights for the year 2008 obtained from the Bureau of Transportation Statistics (see [1]). The database contains flight information and on-time performance statistics for domestic flights; for this example we have selected 500 flights at random from January 2008 and chosen a small subset of the available attributes for each flight as part of the data. The data is stored in the flight_data_subset.csv.

**Step 2. Load a data table from csv file**

As part of our software distribution we provide utilities for easy ingestion of data from csv files. The utility read_model_data_from_csv, ingests the data, extracts data types for each column and constructs Metadata variables for subsequent use. We also provide plotting utilities which allow the user to plot or save the raw data in the PNG format.

```python
filename = os.path.join(S.path.examples_dir, 'flight_data_subset.csv')
filebase = 'flight_data_subset'
T, M_r, M_c = du.read_model_data_from_csv(filename, gen_seed=0)
T_array = numpy.asarray(T)
num_rows = len(T)
num_cols = len(T[0])
col_names = numpy.array([M_c['idx_to_name'][str(col_idx)] for col_idx in range(num_cols)])
dataplot_filename = '{!s}_data'.format(filebase)
```

We can print the attribute names of each column and the type of model CrossCat has identified for use with each attribute.  The _normal inverse gamma model_ is used for continuous attributes and the _symmetric dirichlet discrete model_ is used for categorical data.

To give you a sense of the data underlying the analysis, we print a few rows of the data table and we plot the entire data table.

**Step 3. Initialize CrossCat Engine and build model**

Once the data is ingested, we build the model using the CrossCat engine. CrossCat uses a Markov Chain Monte Carlo (MCMC) process as its computational engine and therefore we must specify the number of state transitions and chains to be used to estimate the full joint distribution. For this example, in the interests of time and aided by the small size of the dataset, we specify a small number of transitions and chains. In most cases, particularly with larger datasets, one may need to use hundreds of transitions to ensure convergence and tens of chains to ensure a robust solution.

```python
# 3. Initialize CrossCat Engine and Build Model
engine = ccc.get_CrossCatClient('local', seed = 0)
X_L_list = []
X_D_list = []
numChains = 10
num_transitions = 10

for chain_idx in range(numChains):
    print 'Chain {!s}'.format(chain_idx)
    M_c_prime, M_r_prime, X_L, X_D = engine.initialize(M_c, M_r, T)
    X_L_prime, X_D_prime = engine.analyze(M_c, T, X_L, X_D, kernel_list=(),
                                          n_steps=num_transitions)
    X_L_list.append(X_L_prime)
    X_D_list.append(X_D_prime)
```

**Step 4. Visualize a cross-categorization of the dataset in one snapshot of the Markov Chain**

Among CrossCat's key capabilities is the ability to partition datasets by first identifying attributes that are predictively related to another, group these attributes into a single "view" and further partition the rows within a view into categories. Since we are estimating the posterior distribution, such "cross-categorization" may vary with each sample drawn from the posterior. After running the MCMC sampler over multiple chains in the above step, we have obtained **X_L_list**, which contains the latent state at the end of each MCMC chain and **X_D_list**, which is the mapping between the model and data for each latent state. With sufficient chains, these variables contain the full joint posterior distribution.

In the figure below, we show the cross-categorization of data obtained from the latent state at the end of the fourth Markov Chain. CrossCat has divided the data into three "views", each with a different number of attributes. One can see that the arrival and departure delays are grouped together, with the cross-categorization showing that the two are positively correlated. Similarly, scheduled and actual arrival and departure times are in the same view, with departure times later in the day leading to arrival times later in the day. However, a separate category is created for flights with late departure times and early arrival times (possibly flights leave before and arrive after midnight). The third view comprises the flight information such as elapsed time, distance, schedule time of the flight and actual time of the flight. The airline and day of week are also grouped into this view. There may be some weak correlation between these variables and flight time/distance but one must also keep in mind that this is simply one possible cross-categorization from among many and this cross-categorization alone may not be (and does not need to be !) completely accurate.

```python
# 4. Visualize clusters in one sample drawn from the model 
viewplot_filename = '{!s}_view'.format(filebase)
pu.plot_views(T_array, X_D_list[4], X_L_list[4], M_c, filename= viewplot_filename)
```

**Step 5: Construct and plot column dependency matrix**

As indicated above, attributes that are predictively related are often assigned to the same "view" during cross-categorization. By aggregating, over the a number of samples from the posterior, the number of times that a pair of attributes are assigned to the same view, we can estimate the dependency between these attributes.

```python
zplot_filename = '{!s}_feature_z'.format(filebase)
# 5. Construct and plot column dependency matrix
su.do_gen_feature_z(X_L_list, X_D_list, M_c, zplot_filename, filename)
```

The adjacent column dependency matrix shows a pairwise measure of dependence between attributes, sorted by hierarchical clustering. The darker the shading in the matrix, the higher the dependence between the two attributes. CrossCat has determined that Arrival Delays are related to Departure Delays, which is intuitive. Also, the flight arrival times are related to departure times; the later in the day a flight leaves, the later it arrives. CrossCat has also identified that flight times and distance are predictively related to the Carrier (airline) which seems to model the characteristics of the airlines (i.e., whether they are regional, national, short-haul dependent, etc.).
